## 9. Primeras vistas por seller

```python
seller_preview = (
    df.groupby("seller_nickname")
    .agg(
        rows=("seller_nickname", "size"),
        unique_urls=("url", "nunique"),
        categories=("category_name", "nunique"),
        avg_price=("price", "mean"),
        median_price=("price", "median"),
        total_stock=("stock", "sum"),
    )
    .sort_values("rows", ascending=False)
)

seller_preview.head(20)
```

## 11. Notas y próximos pasos

Hallazgos de esta exploración inicial:

- **Granularidad:** el dataset está a nivel ítem/publicación (≈4 ítems por seller en promedio); para segmentar vendedores hay que consolidar el grano a nivel seller.
- **Calidad:** destaca un precio máximo de ~$4,77 mil millones MXN (error de carga), `regular_price` nulo en el 73% (comportamiento esperado) y `price` nulo en 0,82%.
- **Logística:** 5 canales (XD 63%, FBM 17%, DS 13,3%, Otro 5,7%, FLEX <1%) → `pct_ds` es relevante.
- **Features candidatas:** precio (log), volumen (ítems/stock), proporciones logísticas, reputación (con flag `has_reputation`) y descuentos.
